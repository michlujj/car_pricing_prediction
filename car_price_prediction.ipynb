```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# to define X and y
X = df.drop('price', axis=1)
```

```python
# to convert all categorical variables into dummy variables
X = pd.get_dummies(X, drop_first=True)
```

```python
y = df['price']
```

```python
# to check for presence of dummy variables
X
```

```python
# testing set= 20%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

```python
scaler = StandardScaler()
```

```python
# to scale the X training and test data to prevent data leakage to unseen data
scaled_X_train = scaler.fit_transform(X_train)
```

```python
scaled_X_test = scaler.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor()
```

```python
# to fit decision tree model to X and y training data
dt.fit(scaled_X_train,y_train)
```

```python
# Decision tree prediction model
dt_pred = dt.predict(scaled_X_test)
```

```python
pd.DataFrame(dt_pred)
```

```python
from sklearn.metrics import r2_score
```

```python
testing_score = []
```

```python
b = r2_score(y_test,dt_pred)*100
testing_score.append(b)
    
print(f"r2_Score of {dt} model on Testing Data is:",b)
```

```python
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
```

```python
rf = RandomForestRegressor()
```

```python
rf.fit(scaled_X_train,y_train)
```

```python
rf_pred = rf.predict(scaled_X_test)
```

```python
rf_pred
```

```python
c = r2_score(y_test,rf_pred)*100
testing_score.append(c)
    
print(f"r2_Score of {rf} model on Testing Data is:",c)
```

```python
ada_model = AdaBoostRegressor()
```

```python
ada_model.fit(scaled_X_train,y_train)
```

```python
ada_model_pred= ada_model.predict(scaled_X_test)
```

```python
ada_model_pred
```

```python
d = r2_score(y_test,ada_model_pred)*100
testing_score.append(d)
    
print(f"r2_Score of {ada_model} model on Testing Data is:",d)
```

```python
gb_model = GradientBoostingRegressor()
```

```python
gb_model.fit(scaled_X_train,y_train)
```

```python
gb_model_pred = gb_model.predict(scaled_X_test)
```

```python
gb_model_pred
```

```python
e = r2_score(y_test,gb_model_pred)*100
testing_score.append(e)
    
print(f"r2_Score of {gb_model} model on Testing Data is:",e) # Accuracy rate of Gradient Boosting regressor highest = 91.29%
```

```python
# to print feature importance of the Gradient Boost regressor model
importances_gb = gb_model.feature_importances_
```

```python
X.columns
```

```python
gb_feat = pd.DataFrame(index=X.columns, data=importances_gb, columns=['Importance'])
```

```python
gb_feat = gb_feat.sort_values('Importance')
```

```python
gb_feat
```

```python
# to remove feature that are of 0 importance, conditional filtering
imp_feat_gb = gb_feat[gb_feat['Importance'] > 0.001]
```

```python
# the sorted Feature Importance model for Gradient Boost model
imp_feat_gb
```

```python
# # Accuracy rate of Gradient Boosting regressor highest = 91.29%
# to visualise in a barplot
# Top 3 predictors of car price: 'enginesize','curbweight','horsepower'
plt.figure(figsize=(10,8),dpi=100)
sns.barplot(data= imp_feat_gb, x='Importance', y=imp_feat_gb.index)
plt.title('Gradient Boost Regressor, feature importance');
```
