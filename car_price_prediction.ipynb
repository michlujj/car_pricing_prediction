```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# to define X and y
X = df.drop('price', axis=1)
```

```python
# to convert all categorical variables into dummy variables
X = pd.get_dummies(X, drop_first=True)
```

```python
y = df['price']
```

```python
# to check for presence of dummy variables
X
```

```python
# testing set= 20%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
scaler = StandardScaler()
```

```python
# to scale the X training and test data to prevent data leakage to unseen data
scaled_X_train = scaler.fit_transform(X_train)
```

```python
scaled_X_test = scaler.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor()
```

```python
# to fit decision tree model to X and y training data
dt.fit(scaled_X_train,y_train)
```

```python
# Decision tree prediction model
dt_pred = dt.predict(scaled_X_test)
dt_pred[:10]
```

```python
# to predict on Decision tree training set
dt_train_pred = dt.predict(scaled_X_train)
dt_train_pred[:10]
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
# MSE for Decision Tree regressor, for testing data
mean_squared_error(y_test,dt_pred)
```

```python
# MSE for Decision Tree regressor, for training data
mean_squared_error(y_train, dt_train_pred)
```

```python
print('Decision Tree_test_RMSE:', np.sqrt(mean_squared_error(y_test, dt_pred)))   # Decision Tree test RMSE = 1974.5
print('Decision Tree_train_RMSE:', np.sqrt(mean_squared_error(y_train, dt_train_pred)))
```
